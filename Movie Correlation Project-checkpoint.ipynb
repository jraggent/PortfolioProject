```python
# import the packages used in this project
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Now read in the data
df = pd.read_csv(r'C:\Users\John\Downloads\archive\movies.csv')
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Taking a look at the data

df.head()
```

```python
# Loop to look for missing data 
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
#Drop rows with any null values
df = df.dropna()
```

```python
# Data Types for our columns

print(df.dtypes)
```

```python
# Look for outliers

df.boxplot(column=['gross'])
```

```python
df.drop_duplicates()
df.head()
```

```python
# Order our Data a little bit to see

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Correlation Matrix between all numeric columns

df.corr(method ='pearson')
```

```python
correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
# Using factorize - assigns a numeric value to every value to find correlation between dataframes

df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

corr_pairs.head()
```

```python
# Scatter plot of Budget vs the Gross Earnings
plt.scatter(x=df['budget'], y=df['gross'], alpha=0.5)
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
```

```python
# Using cat.codes - assigns a numeric value to columns with data type 'object' and 'category' 
df_numerized = df


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
df_numerized.corr(method='pearson')
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```
